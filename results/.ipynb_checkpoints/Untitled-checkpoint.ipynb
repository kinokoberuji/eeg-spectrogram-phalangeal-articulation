```python
#!/usr/bin/env python

from json import load

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

with open("results.json", "r") as results_file:
    results = load(results_file)

# replication
# training accuracies
trn_acc = pd.DataFrame({trial: metric_values 
                        for trial, metrics in results["Replication, 10x sample count"].items()
                        for metric, metric_values in metrics.items()
                        if metric == "categorical_accuracy"})
trn_acc["Average"] = trn_acc.mean(axis=1)

trn_acc
```

```python
# validation accuracies
val_acc = pd.DataFrame({trial: metric_values 
                        for trial, metrics in results["Replication, 10x sample count"].items()
                        for metric, metric_values in metrics.items()
                        if metric == "val_categorical_accuracy"})
val_acc["Average"] = val_acc.mean(axis=1)

val_acc
```

```python
# plot averages
epochs = range(1,11)
fig, ax = plt.subplots()
ax.plot(epochs, trn_acc["Average"], "bo", label="Training")
ax.plot(epochs, val_acc["Average"], "gx", label="Validation")
ax.set_title("Replication, 10x sample count average accuracies")
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracies")
ax.legend(loc="lower right")
plt.show()
```
